```python
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/ML/11.project2/text_categorization/data/yektanet_train.csv')
data
```

```python
# splitting the dataset
train, test = train_test_split(data, test_size = 0.2)
```

```python
# applyingoversampler = RandomOverSampler()
oversampler = RandomOverSampler()
X_train, y_train = oversampler.fit_resample(train, train['category'])
```

```python
X_train
```

```python
# function to preprocess the text
normalizer = Normalizer()
stemmer = Stemmer()
stopwords = stopwords_list()

def preprocessing(text):

    # 1. remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # 2. normalization
    normalized_text = normalizer.normalize(text)

    # 3. apply stemming
    stemmed_text = ' '.join([stemmer.stem(word) for word in normalized_text.split()])

    # 4. tokenization
    tokens = word_tokenize(stemmed_text)

    # 5. convert to lowercase, remove digits and stopwords
    stopwords = stopwords_list()
    output_tokens = []
    for token in tokens:
        token = token.lower()
        if token.isdigit() or token in stopwords:
            continue
        output_tokens.append(token)

    return output_tokens
```

```python
# applying the preprocessing function to the 'title' column in both training and test data
X_train = X_train['title'].apply(preprocessing)
X_test = test['title'].apply(preprocessing)
```

```python
# converting the preprocessed 'title' columns from lists to strings and then apply TF-IDF vectorization to both training and test data
tfidf = TfidfVectorizer()

X_train = X_train.apply(lambda x: " ".join(x) if isinstance(x, list) else x)
X_test = X_test.apply(lambda x: " ".join(x) if isinstance(x, list) else x)

X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
```

```python
train['category'].unique()
```

```python
# encoding the target labels ('y_train') using LabelEncoder to convert them into numeric format
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(test['category'])
```

```python
# training an SVM model with a linear kernel on the training data
model = SVC(kernel='linear', max_iter=5000)
model.fit(X_train, y_train)
```

```python
# predicting on the train and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# model evaluation
score_1 = f1_score(y_train, y_train_pred, average='weighted')
print(f"f1 score on Training Data: {score_1}")

score_2 = f1_score(y_test, y_test_pred, average='weighted')
print(f"f1 score on Training Data: {score_2}")
```
